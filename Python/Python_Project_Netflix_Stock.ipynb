```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns

from datetime import datetime                           #used for Clear data time format
```

```python
df=pd.read_csv("NFLX.csv")
df
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df=df.set_index("Date")
df.head()
```

```python
sns.lineplot(x=df.index,y=df["Volume"],label="Volume")
plt.title("Volume of Stock Vs Year")
```

```python
df.plot(y=["High","Close","Open"],title="Netflix Stock Price")
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(3,figsize = (15,15))
df.groupby(df.index.day).mean().plot(y="Volume",ax=ax1,xlabel="Day")
df.groupby(df.index.month).mean().plot(y="Volume",ax=ax2,xlabel="Month")
df.groupby(df.index.year).mean().plot(y="Volume",ax=ax3,xlabel="Year")
```

```python
a=df.sort_values(by="High",ascending=False).head(5)
a["High"]
```

```python
b=df.sort_values(by="Low",ascending =True).head(5)
b["Low"]
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,sharex=True,figsize=(16,5))
fig.suptitle("High & Low Stock Value As period of Time",fontsize=22)
sns.lineplot(ax=axes[0],x=df.index,y=df["High"],color="green")

sns.lineplot(ax=axes[1],x=df.index,y=df["Low"],color="red")
```
