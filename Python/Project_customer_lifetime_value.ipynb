```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data=pd.read_csv("customer_acquisition_data.csv")
data
```

```python
data.head()
```

```python
fig=px.histogram(data,x="cost",title="Distribution of Customer Acquisition Cost Analysis",nbins=20)
fig.show()
```

```python
fig1=px.histogram(data,x="revenue",title="Revenue Generated by Customer",nbins=20)
fig1.show()
```

```python
data.head()
```

```python
channel_cost=data.groupby("channel")["cost"].mean().reset_index()              #1st Way
channel_cost

fig2=px.pie(channel_cost,values="cost",names="channel",hole=0.4)
fig2.update_layout(title_text="Cost of Channels",title_font=dict(size=20))
fig2.update_traces(textinfo="label+percent",textposition="inside",marker=dict(line=dict(color="black",width=1.2)))

fig2.show()
```

```python
channel_cost2=data.groupby("channel")["cost"].mean().reset_index()              #2nd Way
channel_cost2
colors=["blue","red","lightgreen","yellow"]

fig3=px.bar(channel_cost2,x="channel",y="cost",title="Cost of Channels",color="channel")

fig3.show()
```

```python
channel_conversion=data.groupby("channel")["conversion_rate"].mean().reset_index()              
channel_conversion

fig4=px.bar(channel_conversion,x="channel",y="conversion_rate",title="Conveting Rate of Channels",color="channel")

fig4
```

```python
channel_revenue=data.groupby("channel")["revenue"].sum().reset_index()                  #1st Way
channel_revenue

fig5=px.bar(channel_revenue,x="channel",y="revenue",title="Total Revenue Generated by Channels",color="channel")

fig5.show()
```

```python
channel_revenue2=data.groupby("channel")["revenue"].sum().reset_index()                  #2nd Way
channel_revenue2
colors=["cyan","red","lightgreen","yellow"]

fig6=px.pie(channel_revenue2,values="revenue",names="channel",hole=0.4)
fig6.update_layout(title_text="Total Revenue Generated by Channels",title_font=dict(size=20))
fig6.update_traces(textposition="inside",textinfo="label+percent",marker=dict(colors=colors,
                                                                              line=dict(color="black",width=1.5)))

fig6.show()
```

```python
data.head()
```

```python
data["ROI"]=data["revenue"]/data["cost"]
#data.head()
roi_of_channel=data.groupby("channel")["ROI"].mean().reset_index()
roi_of_channel

fig7=px.bar(roi_of_channel,x="channel",y="ROI",title="Average Return On Investment of Channels",color="channel")

fig7.show()
```
